is_apartment - можем предположить, что "нет данных" там, где ответ - "0", т.е. это не апартаменты. Причиной может быть следующее: не многие люди могли знать разницу между апартаментами и квартирами, это вело к тому, что люди, которые продавали квартиры, при заполнении формы для создания объявления могли просто пропускать эти пункты, т.к. особо не понимали, о чем их спрашивали.
balcony - вполне возможно, что где балкон отсутствует, там вместо 0 стоит NaN. Необходимо проверить, действительно ли по столбцу balcony присутствует только значение "1", говорящее, что балкон есть, а "0" - отсутствует; 
ceiling_height - довольно важная информация, которая отсуствует. Предполагаю, что некоторые не считают это обязательным для заполнения пунктом и поэтому не предприняли никаких действий для того, чтоб узнать высоту своих потолков, иначе бы данные присутствовали. Будем смотреть, как можно заполнить. Считаю, что пропуски можно заполнить медианным значением.
days_exposition - Пропуски могли быть допущены, по той причине, что квартира еще продается. Или же люди продающие квартриу не отобразили, о закрытие сделки.
parks_nearest, ponds_nearest - вполне возможно, что данные просто не были выгружены системой. Беря в расчет то, что это далеко не самые важные параметры, и отсутствие возможности проставить какое-либо адекватные расстояние, которое не испортит статистику по уже присутствующим данным по этим столбцам, думаю, здесь можем не трогать пропуски, учитывая, что есть достаточное количество строк, где данные присутствуют, и по ним можно сделать срез и заметить определенные тенденции; 
По airports_nearest, cityCenters_nearest: данные рассчитывала система. Они могут отустствовать по причине того, что относительно конкретно взятого населенного пункта/точки искались расстояние в определенном радиусе. Если в пределах этого радиуса ни аэропорта, ни парков не было, то и указать в этом столбце системе было нечего. Как бы то ни было, где указаны названия населенных пунктов (кроме СПб, т.к. это крупный город), найдем их координаты и посчитаем вручную расстояние между ними и аэропортом и центром СПб;
kitchen_area, living_area - данные, которые могут быть не заполнены просто по той причине, что они необязательны к заполнению, а высчитывать жилую площадь/площадь кухни продавцам некогда. Также выясним, нет ли такой связи, что если квартира - студия, то там отсутствуют эти данные. Ведь в них нет явного деления площади на ту, которая относится к кухне, и ту, которая относится к жилой площади;
locality_name - выглядит довольно странно, когда не указывают населенный пункт. Предположим, что свойственно это тем продавцам, которые уверены, что всем очевидно, что если они выставляют квартиру на продажу в разделе сайта "Санкт-Петербург", то и квартира в Санкт-Петербурге, не подозревая даже, что это может быть и Лен. область. Тем не менее, заполним данные только по тем строкам, где указано расстояние до центра. От центра до окраин города в среднем 17 км расстояния, те квартиры, которые находятся в пределах 17 км от центра, получат "Санкт-Петерубург" вместо пропусков.




```python
#Приводим целочисленный тип количество балконов. Заменяем пропущенные значения на ноль.
data['balcony'] = data['balcony'].fillna(0).astype(int)
# Приводим к типу даты дату публикации
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'],format = '%Y-%m-%d')
#Заменяем пропущенные значений "False"
data['is_apartment'] = data['is_apartment'].fillna(False)
# Приводим в целочисленный тип стоимость квартиры
data['last_price'] = data['last_price'].astype(int)
# Приводим в целочисленный тип количество парков. Заменяем пропущенные значения на 0
data['parks_around3000'] = data['parks_around3000'].astype('Int64')
# Приводим в целочисленный тип количество прудов. Заменяем пропущенные значения на 0
data['ponds_around3000'] = data['ponds_around3000'].astype('Int64')
```

```python
data.dropna(subset = ['locality_name'], inplace = True)
data['locality_name'] = data['locality_name'].str.replace('ё', 'е') 
#меняем "посёлок" на "поселок"  (устранение неявных дубликатов)
#удаляем отсутствивующие значения в столбце locality_name, т.к. их мало и они не повлияют на дальнейшее исследование
```

```python
#Заполнение пропусков медианным значением 
data['ceiling_height'].fillna(data['ceiling_height'].median(), inplace=True)
```

```python
#Осуществляем проверку
data['ceiling_height'].describe()
```

```python
#для начала вычислим среднее значение по высоте потолков, ограничив выборку допустимыми значениями
data.query('ceiling_height > 2 and ceiling_height < 8.3')['ceiling_height'].mean()
data.loc[data['ceiling_height'] < 2, 'ceiling_height'] = 2.72 #приравняем невошедшие значения к среднему, т.к. потолков выше 8,3 м. и ниже 2 м. быть не может
data.loc[data['ceiling_height'] > 8.3, 'ceiling_height'] = 2.72
data['ceiling_height'] = data['ceiling_height'].fillna(2.72)
```

```python
data = data.query('rooms < 6')
data = data.reset_index(drop=True)
```

```python
 # Добавляем столбец со стоимостью квадратного метра ('price_per_square_metter') и приводим его в целочисленный тип
data['price_per_square_metter'] = data['last_price'] / data['total_area']
data['price_per_square_metter'] = data['price_per_square_metter'].astype(int)
# Добавляем столбцы дня недели, месяца и года
data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
data['month_exposition'] = data['first_day_exposition'].dt.month
data['year_exposition'] = data['first_day_exposition'].dt.year
# Создаём функцию для заполнения нового столбца с типом этажа
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'Первый'
    elif floor == floors_total:
        return 'последний'
    elif 1 < floor < floors_total:
        return 'другой'
data['floor_category'] = data.apply(floor_category, axis = 1)
#Добавляем расстояние от центра города в КМ
data['distance'] = round(data['city_centers_nearest'] / 1000)
data.head(5)
```
